# Kütüphaneleri İçeri Aktarma

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

# Veri Setinin İçeri Aktarılması

```python
df = pd.read_csv("polynomial.csv", sep =";")
df
```

# Veri Setinin Görsel olarak İncelenmesi

```python
plt.scatter(df["deneyim"],df["maas"])
plt.xlabel("Deneyim (yıl)")
plt.ylabel("Maaş")
plt.savefig("1.png", dpi=300)
plt.show()
```

# Veri Setinin Linear Regresyon ile Değerlendirilmesi

```python
reg = LinearRegression()
reg.fit(df[["deneyim"]],df["maas"])

plt.xlabel("Deneyim (yıl)")
plt.ylabel("Maaş")

plt.scatter(df["deneyim"],df["maas"])

xekseni = df["deneyim"]
yekseni =reg.predict(df[["deneyim"]])
plt.plot(xekseni,yekseni,color = "green", label = "linear regression")
plt.legend()
plt.show()
```

# Veri Setinin Polynomial Regression ile Farklı Derecelerde Değerlendirilmesi

2. Dereceden Polinom

```python
polynomial_regression = PolynomialFeatures(degree = 2)

x_polynomial = polynomial_regression.fit_transform(df[["deneyim"]])

reg = LinearRegression()
reg.fit(x_polynomial,df["maas"])

y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"],y_head, color = "red", label = "polynomial_regression")
plt.legend()

plt.scatter(df["deneyim"],df["maas"])

plt.show()
```

3. Dereceden Polinom

```python
polynomial_regression = PolynomialFeatures(degree = 3)

x_polynomial = polynomial_regression.fit_transform(df[["deneyim"]])

reg = LinearRegression()
reg.fit(x_polynomial,df["maas"])

y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"],y_head, color = "red", label = "polynomial_regression")
plt.legend()

plt.scatter(df["deneyim"],df["maas"])

plt.show()
```
